### Step 2: Import Libraries and Load Data
Now we import the tools we need (like pandas for data and seaborn for graphs) and load our CSV dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style="whitegrid")

# Load the dataset
df = pd.read_csv('../data/SupplyChainSentinel_Enterprise_100000_Fixed_2(Sheet1).csv')
print(f"Dataset Shape: {df.shape}")

# Show the first 5 rows
df.head()
```

### Step 3: Check Data Information and Missing Values
Let's see what types of data we have and if there are any missing values in our dataset.

```python
info_df = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df)) * 100
})
info_df.sort_values(by='Missing %', ascending=False)
```

### Step 4: Summary Statistics
Here we look at the mathematical summary (mean, min, max) for numerical and categorical columns.

```python
# Summary for Numerical features
df.describe()
```

```python
# Summary for Categorical features
df.describe(include=['O'])
```

### Step 5: Univariate Analysis - Delay Days Distribution
Let's create a histogram to see the most common delay times.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['delay_days'].dropna(), bins=30, kde=True, color='coral')
plt.title('Distribution of Delay Days')
plt.xlabel('Delay Days')
plt.ylabel('Frequency')
plt.show()
```

### Step 6: Bivariate Analysis - Material Type vs Delay
Let's see if the type of material affects how many days it gets delayed.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='material', y='delay_days', data=df, palette='Set2')
plt.title('Delay Days by Material Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 7: Bivariate Analysis - Weather vs Delay
Let's check if bad weather causes more delays.

```python
if 'weather' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='weather', y='delay_days', data=df, palette='Set3')
    plt.title('Delay Days by Weather Conditions')
    plt.tight_layout()
    plt.show()
```

### Step 8: Correlation Heatmap
Finally, a heatmap to see how numerical features correlate with each other (e.g. distance vs delay).

```python
plt.figure(figsize=(14, 10))
numerical_df = df.select_dtypes(include=[np.number])
sns.heatmap(numerical_df.corr(), annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
